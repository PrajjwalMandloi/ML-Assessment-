```python
import numpy as np
```

```python
from sklearn.metrics import *
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,mean_squared_error
```

```python
y_true = np.array([100,150,200,250,125,230,255,220])
y_pred = np.array([110,140,210,240,115,260,220,255])
```

```python
#MAE

mae = mean_absolute_error(y_true,y_pred)
print(mae)

#Definition: Average error between actual and predicted values

#Meaning:On average, the predicted values differ from the actual values by about 18.75 units.
```

```python
#MSE

mse = mean_squared_error(y_true,y_pred)
print(mse)

#Definition: MSE measures average squared prediction errors. Because errors are squared, large mistakes are penalized more.

#Meaning: Some errors are larger because squaring increases bigger mistakes
```

```python
#RMSE

rmse = root_mean_squared_error(y_true,y_pred)
print(rmse)

#RMSE (21.9) is close to MAE (18.75) means errors are consistent, not too many big mistakes.
```

```python
#r2 score

r2 = r2_score(y_true,y_pred)
print(r2)

#Definition: How much of the data is explained by the model

#Meaning: The model explains 84% of the data → good performance
```

```python
#adjusted r2 score

n = len(y_true)
p = 1
adj_r2 = 1-(1-r2)*(n-1)/(n-p-1) #p = number of independent variables (features)
print(adj_r2)

# Definition: Adjusted R² corrects R² for number of predictors.Adding more variables always increases R²

# Meaning: Slightly lower than R² → most features are useful (no major overfitting)
```

```python
y_mean = np.mean(y_true)
sst = np.sum((y_true - y_mean)** 2) #sum of (actual-mean)^2
ssr = np.sum((y_true - y_pred)** 2) #(sum of (actual - predicted)^2
print("SST:",sst) #Total Sum of Squares
print("SSR:",ssr) #Sum of Squared Residuals

#(SST)
#Definition: Total variance present in the actual data.
#Interpretation: This represents the overall spread of the data.
#(SSR)
#Definition: Total error made by the model
#Meaning: Error is low compared to total variation → good fit.
```
